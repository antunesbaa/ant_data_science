# Panorama do COVID-19 no Brasil

A COVID-19 é uma doença infecciosa causada pelo coronavírus, transmitida principalmente por gotículas de pessoas infectadas. A gravidade dos sintomas varia de pessoa para pessoa e ainda há muitas incertezas sobre a doença. 

Estudos estão sendo realizados em todo o mundo, mas os resultados ainda não são conclusivos. 

3877986.jpg

Até o momento, 80% dos casos confirmados são assintomáticos e a maioria se recupera sem sequelas. No entanto, 15% dos casos requerem tratamento com oxigênio e 5% precisam de ventilação assistida em ambiente hospitalar. 

Para aumentar a conscientização sobre a COVID-19 no Brasil, realizarei uma análise dos dados públicos disponíveis sobre a doença. Verificar a situação atual da doença é fundamental para o planejamento de políticas públicas e a prevenção de novos casos.

## Obtenção dos Dados

A fonte de dados utilizada para análise da COVID-19 no Brasil é o Ministério da Saúde. O Ministério disponibiliza diariamente um painel atualizado com dados referentes aos casos confirmados, mortes, recuperados, entre outras informações relevantes. O dataset é composto por informações de todo o mundo, desde o início da pandemia em 2020.

Esse dataset pode ser usado para entender a evolução da pandemia, identificar tendências e padrões, e auxiliar na tomada de decisões em saúde pública. No entanto, é importante lembrar que o dataset é constantemente atualizado, e a interpretação dos dados deve ser realizada com cautela, considerando sempre a possibilidade de mudanças nos números.

A análise de uma doença nova, como é o caso da COVID-19, pode apresentar diversas dificuldades. Ainda que a ciência tenha avançado significativamente nos últimos anos, a compreensão completa da doença é algo que ainda está em desenvolvimento. Isso significa que as informações disponíveis estão em constante evolução, e podem ser atualizadas a qualquer momento.

* #### **Verificando os locais que têm mais casos**

```python
df_filtered = df.query("location not in ['World', 'High income', 'Europe', 'Asia', 'Upper middle income', \
'European Union', 'North America', 'Lower middle income', 'South America']") # filtrando para que aparecçam somente países

df_cases = df_filtered.loc[df_filtered.date == '2023-03-07', ['location', 'total_cases']].sort_values(by="total_cases", ascending=False).head(15)

print(df_cases.to_string(index=False))
```

* #### **Verificando os locais que têm mais mortes**

```python
df_filtered = df.query("location not in ['World', 'High income', 'Europe', 'Asia', 'Upper middle income', \
'European Union', 'North America', 'Lower middle income', 'South America']") # filtrando para que aparecçam somente países
df_death = df_filtered.loc[df_filtered.date == '2023-03-07', ['location', 'total_deaths']].sort_values(by="total_deaths", ascending=False).head(15)

print(df_death.to_string(index=False))
```

##### **Em relação ao número de mortes por Covid 19:**


* **Estados Unidos**: 1 milhão 122 mil

##### **Para afcilitar a visualização dos dados, vamos plotar um gráfico de barras horizontal**:

```python
# Selecionando os dados da data especificada e das colunas "location" e "total_deaths", ordenando por "total_deaths" de forma decrescente e selecionando as cinco maiores
df.loc[df.date == '2023-03-07', ['location', 'total_deaths']].sort_values(by="total_deaths", ascending=False)[1:6]

# Criando o gráfico de barras horizontais com a localização no eixo x, o total de mortes no eixo y, a cor verde e sem legenda
ax = df_death.plot.barh(x='location', y='total_deaths', color='green', legend=None)

# Definindo o título do gráfico, o label do eixo x e o label do eixo y
ax.set_title('Total de mortes por COVID-19 até 07-03-2023')
ax.set_xlabel('Total de mortes')
ax.set_ylabel('Localização')

# Invertendo a ordem do eixo y para que as localizações com maior número de mortes fiquem no topo do gráfico
ax.invert_yaxis()

# Mostrando o gráfico
plt.show()
```

* #### **Evolução do total de casos de COVID-19 no mundo**:

Saber sobre a evolução da COVID-19 no mundo é fundamental para entender o impacto da pandemia e tomar decisões informadas. Uma das melhores formas de visualizar essa evolução é através do gráfico de linhas, que permite acompanhar a tendência de casos ao longo do tempo. 

```python
# agrupar os dados por data e soma as mortes totais para todo o mundo
df_world = df.groupby('date')['total_deaths'].sum()

# plotar o gráfico de linha
plt.plot(df_world.index, df_world)

# definir o título e os rótulos dos eixos
plt.title('Evolução do número total de mortes por COVID-19 no mundo')
plt.xlabel('Data')
plt.ylabel('Número total de mortes')

# exibir o gráfico
plt.show()
```
